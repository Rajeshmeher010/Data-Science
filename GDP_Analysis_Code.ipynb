```python
#Part-Ia: GDP Analysis of Indian States

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import figure

#Read data from input file.
df = pd.read_csv("C:\\Users\Rajesh.Meher\Desktop\DataScience\Project\gdp_analysis\part_1\gdp_analysis.csv")
df = df.drop('West Bengal1',axis = 1)

#Remove the below rows for the year 2016-17.
df = df.drop([5,10])
df1 = df.iloc[6:10]
df1 = df1.drop(['Items  Description'],axis=1,inplace = False)

#Calculate the average growth of states over the duration 2013-14, 2014-15 and 2015-16
df1 = df1.T
df1 = df1.drop('Duration',axis = 0)
df1 = df1.rename(columns ={7:2013,8:2014,9:2015},inplace = False)
df1["Mean"] = df1.mean(axis=1,skipna = True)
df1 =  df1.sort_values('Mean',inplace=False,ascending=False)
plt.figure(figsize= (13,10))
sns.barplot(x = df1['Mean'],y = df1.index,data = df1)
plt.show()
```
